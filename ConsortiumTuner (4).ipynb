## Step 3: The scoring engine

Given slider weights, score each circuit, pick the ones above the selection threshold, and then add any required dependencies (e.g. you can't run KillSwitch_StrainA without also having StrainA_Gsulfurreducens).

```python
def score_circuits(weights, threshold=0.25):
    """
    Score each circuit by dotting its score vector with the user's slider weights.
    Equal-weighted: each circuit scores 1.0 on its category, 0.0 elsewhere.
    Returns a dict: circuit_name -> score.
    """
    scored = {}
    for name, info in CIRCUITS.items():
        total = sum(weights[axis] * info['scores'][axis] for axis in weights)
        scored[name] = total
    return scored


def select_consortium(weights, threshold=0.25):
    """
    1. Score all circuits.
    2. Select those above threshold.
    3. Enforce dependencies (e.g. if KillSwitch_A is selected, StrainA must be too).
    """
    scores = score_circuits(weights)
    selected = {name for name, s in scores.items() if s >= threshold}

    # Dependency closure
    changed = True
    while changed:
        changed = False
        for name in list(selected):
            for req in CIRCUITS[name]['requires']:
                if req not in selected:
                    selected.add(req)
                    changed = True

    return sorted(selected, key=lambda n: -scores[n]), scores


# Quick test: set all sliders to 0.5 — should select circuits from every category that's >= threshold
test_weights = {
    'cross_feeding': 0.5, 'kill_switch': 0.5, 'electroactivity': 0.5,
    'nitrification': 0.5, 'denitrification': 0.5, 'production': 0.5, 'quorum_sensing': 0.5
}
selected, scores = select_consortium(test_weights)
print(f"Test selection ({len(selected)} circuits, all sliders = 0.5):")
for name in selected:
    print(f"  {scores[name]:.2f}  {name}")
```

## Step 4: The visualization

When the user moves a slider, render:
1. A **ranked list** of selected circuits with their GOLDBAR specs
2. A **grid of Knox screenshots** for the selected circuits (the actual design spaces)
3. A **scoring bar chart** showing which circuits cleared the threshold
4. An **ecosystem summary** of the organisms in the final consortium

```python
# Color palette (extended for 7 categories)
CATEGORY_COLORS = {
    'cross_feeding':   '#8FD14F',
    'kill_switch':     '#F59E0B',
    'electroactivity': '#00B4B4',
    'nitrification':   '#A855F7',
    'denitrification': '#EC4899',
    'production':      '#EF4444',
    'quorum_sensing':  '#3B82F6',
}


def render_recommendation(weights, threshold=0.25):
    """Render everything: list, image grid, bar chart, summary."""
    selected, scores = select_consortium(weights, threshold)

    if not selected:
        display(Markdown("### ⚠️ No circuits cleared the threshold — try raising some sliders."))
        return

    # ----- 1. Recommended circuits with GOLDBAR specs -----
    md = [f"## 🧬 Your consortium — {len(selected)} circuits\n"]
    md.append("| Score | Circuit | Organism | GOLDBAR |")
    md.append("|---|---|---|---|")
    for name in selected:
        info = CIRCUITS[name]
        md.append(f"| **{scores[name]:.2f}** | `{name}` | *{info['organism']}* | `{info['spec']}` |")
    display(Markdown('\n'.join(md)))

    # ----- 2. Image grid of Knox design spaces -----
    if HAS_IMAGES and selected:
        n = len(selected)
        cols = min(3, n)
        rows = (n + cols - 1) // cols
        fig, axes = plt.subplots(rows, cols, figsize=(5.5 * cols, 4.5 * rows))
        fig.suptitle('Knox design spaces for the selected consortium', fontsize=14, y=1.00)
        if rows * cols == 1:
            axes = np.array([[axes]])
        elif rows == 1:
            axes = axes.reshape(1, -1)
        elif cols == 1:
            axes = axes.reshape(-1, 1)

        for i, name in enumerate(selected):
            r, c = i // cols, i % cols
            ax = axes[r, c]
            info = CIRCUITS[name]
            img_path = os.path.join(IMAGES_DIR, info['image'])
            if os.path.exists(img_path):
                img = mpimg.imread(img_path)
                ax.imshow(img)
            else:
                ax.text(0.5, 0.5, '(image not found)', ha='center', va='center', transform=ax.transAxes)
            color = CATEGORY_COLORS[info['category']]
            ax.set_title(name, fontsize=11, fontweight='bold', color=color, pad=8)
            ax.axis('off')
            for spine in ax.spines.values():
                spine.set_visible(True)
                spine.set_edgecolor(color)
                spine.set_linewidth(2)

        for i in range(n, rows * cols):
            r, c = i // cols, i % cols
            axes[r, c].axis('off')

        plt.tight_layout()
        plt.show()

    # ----- 3. Score bar chart -----
    fig, ax = plt.subplots(figsize=(11, 7))
    all_names = list(CIRCUITS.keys())
    all_scores = [scores[n] for n in all_names]
    colors = [CATEGORY_COLORS[CIRCUITS[n]['category']] if n in selected else '#D0D9E0' for n in all_names]
    edge_colors = ['#0B1F2A' if n in selected else '#D0D9E0' for n in all_names]
    bars = ax.barh(all_names, all_scores, color=colors, edgecolor=edge_colors, linewidth=1.5)
    ax.axvline(threshold, color='#EF4444', linestyle='--', linewidth=1.5, label=f'Selection threshold = {threshold}')
    ax.set_xlabel('Fitness score (weight on active axis, equal-weighted)', fontsize=11)
    ax.set_xlim(0, 1.05)
    ax.invert_yaxis()
    ax.legend(loc='lower right', fontsize=10)
    ax.set_title('Circuit fitness scores — selected circuits in color, rejected in grey', fontsize=12, pad=10)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for bar, score in zip(bars, all_scores):
        if score > 0.02:
            ax.text(score + 0.01, bar.get_y() + bar.get_height()/2, f'{score:.2f}',
                    va='center', fontsize=9, color='#0B1F2A')
    plt.tight_layout()
    plt.show()

    # ----- 4. Ecosystem summary -----
    md2 = ["### 🌍 Ecosystem composition\n"]
    by_cat = {}
    for name in selected:
        cat = CIRCUITS[name]['category']
        by_cat.setdefault(cat, []).append(name)
    for cat in ['cross_feeding', 'kill_switch', 'electroactivity', 'nitrification', 'denitrification', 'production', 'quorum_sensing']:
        if cat in by_cat:
            cat_label = cat.replace('_', '-').title()
            md2.append(f"**{cat_label}** ({len(by_cat[cat])}): " + ", ".join(f"`{n}`" for n in by_cat[cat]))
    organisms = sorted(set(CIRCUITS[n]['organism'] for n in selected))
    md2.append(f"\n**Organisms in consortium** ({len(organisms)}): " + ", ".join(f"*{o}*" for o in organisms))
    display(Markdown('\n'.join(md2)))


print("Rendering function ready.")
```

## What happens next

1. **Export the selected consortium as SBOL** (next milestone) — the selected circuits get written to an SBOL XML file that iBioSim can import.
2. **Simulate in iBioSim** — each circuit becomes a hierarchical submodel, with a shared extracellular medium for cross-feeding and N-cycle metabolites.
3. **Parameter sensitivity sweep** — for any slider setting the user lands on, run an ODE + SSA analysis to predict current output, stability, and effluent composition.
